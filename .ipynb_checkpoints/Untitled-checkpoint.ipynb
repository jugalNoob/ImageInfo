```python
import PIL.Image
import PIL.ExifTags

oneimp = PIL.Image.open("one.jpg")

exif = oneimp._getexif()
if exif is not None:
    exif_data = {
        PIL.ExifTags.TAGS[k]: v
        for k, v in exif.items()
        if k in PIL.ExifTags.TAGS
    }
    
    # Access the specific keys for date and time information
    date_time_original = exif_data.get('DateTimeOriginal')
    date_time_digitized = exif_data.get('DateTimeDigitized')
    Model=exif_data.get('Model')
    Maker=exif_data.get("Make")
    Software=exif_data.get("Software")
    
    print("Original Date and Time:", date_time_original)
    print("Digitized Date and Time:", date_time_digitized)
    print("Model:" , Model)
    print("Maker:" , Maker)
    print("Software:" Software)
    


    
else:
    print("No EXIF metadata available for this image.")
```

```python
north=exif['GPSInfo'][2]
east=exif['GPSInfo'][4]
```

```python
lats=((((north[0] * 60)+north[1])*60)+north[2])/60/60
long=((((east[0] * 60)+north[1])*60)+north[2])/60/60
```

```python
from gmplot import gmplot

gmap=gmplot.GoogleMapPlotter(lats , long , 12)

gmap.marker(lats , long, "cornflowerblue")
gmap.draw("jugal.html")
```

```python

from geopy.geocoders import Nomination

geoloc=Nomination(user_agent="GetLoc")

locname=geoloc.reverse(f"{lats} , {long}")
print(locname.address)
```
